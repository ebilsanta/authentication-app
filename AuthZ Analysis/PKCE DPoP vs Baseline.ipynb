```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statistics import mean
```

Create a dataframe with all the datapoints

```python
df = pd.DataFrame({'ac_base': pd.Series(np.loadtxt('authcode_speed_base.csv', delimiter=',')),
             'ac_dpop_pkce': pd.Series(np.loadtxt('authcode_speed.csv', delimiter=',')),
             'token_base': pd.Series(np.loadtxt('token_speed_base.csv', delimiter=',')),
             'token_dpop_pkce': pd.Series(np.loadtxt('token_speed.csv', delimiter=',')),
             'refresh_base': pd.Series(np.loadtxt('refresh_speed_base.csv', delimiter=',')),
             'refresh_dpop_pkce': pd.Series(np.loadtxt('refresh_speed.csv', delimiter=',')),})

```

Plot bargraph with means and S.Ds

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
fig.tight_layout()

sns.barplot(df[['ac_base', 'ac_dpop_pkce']], ax=axes[0])
sns.barplot(df[['token_base', 'token_dpop_pkce']], ax=axes[1])
sns.barplot(df[['refresh_base', 'refresh_dpop_pkce']], ax=axes[2])
fig.suptitle('Main title')
```

Print the means and S.Ds

```python
print(mean(df['ac_base']))
print(mean(df['ac_dpop_pkce']))
print(mean(df['token_base']))
print(mean(df['token_dpop_pkce']))
print(mean(df['refresh_base']))
print(mean(df['refresh_dpop_pkce']))
```
